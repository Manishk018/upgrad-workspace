```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np, pandas as pd
```

```python
chu_df=pd.read_csv('churn_data.csv')
cus_df=pd.read_csv('customer_data.csv')
int_df=pd.read_csv('internet_data.csv')
```

```python
tempo=pd.merge(chu_df,cus_df,how='inner', on='customerID')
```

```python
df=pd.merge(tempo,int_df,how='inner',on='customerID')
df
```

```python
df['OnlineBackup'].value_counts()
```

```python
cols=['PhoneService', 'PaperlessBilling','Churn','Partner','Dependents']

def binary_map(x):
    return x.map({'Yes':1,'No':0})

df[cols]=df[cols].apply(binary_map)
```

```python
df.head()
```

```python
dumcol=['Contract','PaymentMethod','gender','InternetService']
dummies=pd.get_dummies(df[dumcol],drop_first=True)
dummies.head()
```

```python
df=pd.concat([dummies,df], axis=1)
```
