```python
df = pd.read_csv('https://media-doselect.s3.amazonaws.com/generic/NMgEjwkAEGGQZBoNYGr9Ld7w0/rating.csv')
# print(df)
df['Training'] = df['Rating'].apply(lambda x : 'Yes' if x <= 3.5 else 'No')
print(df)
```

```python
dfgroup = df.groupby(['Department']).count()
dfgroup
```

```python
for i in ['Finance', 'HR', 'Sales', 'Marketing']:
    print(i, len(df[(df['Training'] == 'Yes') & (df['Department'] == i)]) / len(df[df['Department'] == i]) * 100)
```

```python
df.groupby(df['Department']).mean().sort_values('Rating')
```
